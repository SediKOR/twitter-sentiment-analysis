# EDA

```python
from wordcloud import WordCloud #Word visualization
import matplotlib.pyplot as plt #Plotting properties
import pandas as pd
```

```python
train_data = pd.read_csv('./preprocessing_train.csv', index_col=0)
val_data = pd.read_csv('./preprocessing_val.csv', index_col=0)
```

```python
#train_data에서 긍정 부정 유무가 기재된 레이블 값 분포확인
train_data['type'].value_counts().plot(kind = 'bar')
plt.xticks(rotation=45)
```

## WordCloud

```python
word_cloud_text = ''.join(train_data[train_data["type"]==1].lower)
#Creation of wordcloud
wordcloud = WordCloud(
    max_font_size=100,
    max_words=100,
    background_color="white",
    scale=10,
    width=800,
    height=800
).generate(word_cloud_text)
#Figure properties
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
